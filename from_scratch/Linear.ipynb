```python
import numpy as np
import scipy.special 
import matplotlib.pyplot as plt
from NN import NN_Classifier
```

```python
nn = NN_Classifier([3, 3, 3], 0.3)
```

```python
# random data

data = np.random.rand(10000,3)
target = data.copy()
target[:,0] += 2*data[:,1]
target[:,2] += 2*data[:,0] - 2*data[:,2]

training_data = data[:7500,:]; training_target = target[:7500,:]
test_data = data[7500:,:]; test_target = target[7500:,:]
```

```python
# train

EPOCHS = 10; train_loss = []
for e in range(EPOCHS):
    for data, target in zip(training_data,training_target):
        nn.train(data,target)
        train_loss.append(nn.get_loss())
```

```python
# test 
answers = []; predictions = []; test_loss = []
for data, target in zip(test_data,test_target):
    answers.append(target)
    predictions.append(nn.test(data))
    test_loss.append(nn.get_loss(target))
```

```python
i = 0
for z,y in zip(answers, predictions):
    if (i<10):
        print(z,y)
        i+=1
    else:
        break
```

```python
import matplotlib.pyplot as plt 
n=len(test_loss)

plt.plot([x for x in range(n)], train_loss[:n])
plt.plot([x for x in range(n)], test_loss[:n])
plt.legend(labels=['train','test'])
plt.show()
```
